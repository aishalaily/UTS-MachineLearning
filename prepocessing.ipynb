```python
# Import library
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("pldb.csv")

# Tampilkan 5 data teratas
df.head()
# Menampilkan baris index 0 sampai 19 (baris ke-1 sampai 20)
# df.loc[0:19]
```

```python
# Informasi dataset
df.info()

print("Jumlah missing value per kolom:")
# Cek jumlah missing value
print(df.isnull().sum())
```

```python
# Cek persentase missing value tiap kolom
missing_percentage = df.isnull().mean() * 100
print("Persentase missing value per kolom:")
print(missing_percentage)
```

```python
# Hapus kolom yang missing value > 50%
cols_to_drop = missing_percentage[missing_percentage > 50].index
df.drop(columns=cols_to_drop, inplace=True)
```

```python
# Cell 7 - Isi Missing Value (Numerik → Median, Kategorikal → Mode)
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    df[col].fillna(df[col].median(), inplace=True)
for col in df.select_dtypes(include=['object']).columns:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Cek ulang setelah preprocessing

# Hitung kolom yang dihapus
deleted_cols = missing_percentage[missing_percentage > 50].index
print(f"\nJumlah kolom yang dihapus: {len(deleted_cols)}")
print("Kolom yang dihapus:", list(deleted_cols))

# Hitung kolom yang diubah (karena sebelumnya ada missing dan diisi median/mode)
changed_cols = [col for col in df.columns if df[col].isnull().sum() == 0 and col in missing_percentage[missing_percentage > 0].index]
print(f"\nJumlah kolom yang diubah (diisi median/mode): {len(changed_cols)}")
print("Kolom yang diubah:", changed_cols)

# Cek ulang sisa missing value
print("\nSisa missing value setelah preprocessing (per kolom):")
print(df.isnull().sum().sort_values(ascending=False).head(20))

```

```python
# Cell 9 - Simpan dataset yang sudah bersih
df.to_csv("pldb_clean.csv", index=False)
print("\nDataset bersih sudah disimpan ke 'pldb_clean.csv'")
```
